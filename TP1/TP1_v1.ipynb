```python
# show their dimentions
B = cv.imread('B.png')
print(B.shape)
# get the storage size of the image
print(B.nbytes)
# get the size using os
print(os.path.getsize('B.png'))
```

```python

C = cv.imread('C.tif')
print(C.shape)
print("TIF size",C.size)
# get the storage size of the image
print(os.path.getsize('C.tif'))
```

```python
# Compare the three formats
plt.subplot(1,3,1)
plt.imshow(A)
plt.subplot(1,3,2)
plt.imshow(B)
plt.subplot(1,3,3)
plt.imshow(C)
plt.show()
```

```python
# show the levels of the image, red, blue and green
img = cv.imread(path)

r = img.copy()
r[:,:,0] = r[:,:,1] = 0

g = img.copy()
g[:,:,0] = g[:,:,2] = 0

b = img.copy()
b[:,:,1] = b[:,:,2] = 0

cv.imshow("red",r)
cv.imshow("green",g)
cv.imshow("blue",b)
cv.waitKey(0)
cv.destroyAllWindows()

```

```python
# convert the image to grayscale
img = cv.imread(path,cv.IMREAD_GRAYSCALE)
cv.imshow('Grayscale Image', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
thresh_value = 128
max_value = 255
binary_img = cv.threshold(img, thresh_value, max_value, cv.THRESH_BINARY)[1]

# Display the binary image
cv.imshow('Binary Image', binary_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
